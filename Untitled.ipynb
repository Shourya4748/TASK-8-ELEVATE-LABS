```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score

%matplotlib inline
```

```python

df = pd.read_csv('Mall_Customers.csv')
df.head()
```

```python

print(df.info())
print(df.describe())

data = df.drop(columns=['CustomerID', 'Gender']) 
```

```python

pca = PCA(n_components=2)
pca_data = pca.fit_transform(data)
plt.figure(figsize=(8,6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c='gray')
plt.title('PCA Projection (Unclustered)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.grid(True)
plt.show()
```

```python

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(data)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8,6))
plt.plot(K_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
optimal_k = 5 
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans.fit_predict(data)
df['Cluster'] = clusters
```

```python

pca_data_clustered = PCA(n_components=2).fit_transform(data)

plt.figure(figsize=(8,6))
sns.scatterplot(x=pca_data_clustered[:, 0], y=pca_data_clustered[:, 1],
                hue=clusters, palette='Set2', s=100)
plt.title('K-Means Clusters (PCA Reduced)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python

score = silhouette_score(data, clusters)
print(f"Silhouette Score for K={optimal_k}: {score:.3f}")
```
